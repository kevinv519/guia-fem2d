1. Para la ecuación de calor utilizada en clase
$$ \nabla(k \nabla T) = - Q $$
Modificar la formulación del FEM para
$$ Q = Q(x) = 2x^4-3x^2$$

**Desarrollo**
Para la integral
$$ \iint_\Omega \mathbf{N}^\top Q(x) dy dx $$

Definiendo que:
$$ \psi_{ij} = \psi_i - \psi_j ; \psi \in \{x,y\} $$

Donde:
$$ x = x_{21} \epsilon + x_{31} \eta + x_1 $$


```python
from sympy import *
init_printing()

eps, eta, x21, x31, x2, x3, x1 = symbols('\epsilon \eta x_21 x_31 x_2 x_3 x_1')

x = (x2-x1)*eps + (x3-x1)*eta + x1
#x = x21*eps + x31*eta + x1

nt = Matrix([1-eps-eta, eps, eta])


q = (2*x**4 - 3*x**2)
```

```python
b1 = Integral(Integral(nt*q, [eta, 0, 1-eps]), [eps, 0, 1])
b1
```

```python
e = integrate(integrate(nt*q, [eta, 0, 1-eps]), [eps, 0, 1])
e
```

La primera observación que se puede hacer es que ahora, nuestra x está definida en función de las variables isoparamétricas $\{\epsilon, \eta\}$ y por consiguiente, al multiplicarle por el vector columna que representa las funciones de forma transpuesta, nos damos cuenta que el resultado ya no serían los mismo valores, como en la formulación original del método para 2D con $Q$ constante
